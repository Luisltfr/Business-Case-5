# Quarterly Analysis:
## Top Products sold

We start with analysis the top 5 products sold per store per quarter, taking into consideration units sold only. 
We will upload the 4 datasets related to the 4 different years and from there will construct a table per year that will show the top 5 products divided by quarter, together with the total of units sold for that quarter.

```python
df19=pd.read_csv('2019.csv')
df=df19
```

```python
df.info()
```

```python
df['week'] = pd.to_datetime(df['week'])
df['month'] = df['week'].dt.month
df['quarter'] = df['week'].dt.quarter
```

```python
df.head()
```

```python
#For 2019 we will split the dataset in quarters in order to find the top 5 products sold per store per quarter
df1=df[df['quarter']==1]
df2=df[df['quarter']==2]
df3=df[df['quarter']==3]
df4=df[df['quarter']==4]
```

```python
#for the first quarter we select store and name of the product and we calculate the sum of units sold
quarter1 = df1.groupby(['store','Name']).agg({'  u':['sum']})
quarter1.columns = quarter1.columns.droplevel(1)
quarter1 = quarter1.sort_values(by=['store','  u'], ascending=False) #we sort Product names by units sold
quarter1 = quarter1.groupby(level=0).head(5) #we decided to show only the top 5 products for each store based on units sold
```

```python
#for the second quarter we select store and name of the product and we calculate the sum of units sold
quarter2 = df2.groupby(['store','Name']).agg({'  u':['sum']})
quarter2.columns = quarter2.columns.droplevel(1)
quarter2 = quarter2.sort_values(by=['store','  u'], ascending=False) #we sort Product names by units sold
quarter2 = quarter2.groupby(level=0).head(5) #we decided to show only the top 5 products for each store based on units sold
```

```python
#for the second quarter we select store and name of the product and we calculate the sum of units sold
quarter3 = df3.groupby(['store','Name']).agg({'  u':['sum']})
quarter3.columns = quarter3.columns.droplevel(1)
quarter3 = quarter3.sort_values(by=['store','  u'], ascending=False) #we sort Product names by units sold
quarter3 = quarter3.groupby(level=0).head(5) #we decided to show only the top 5 products for each store based on units sold
```

```python
#for the second quarter we select store and name of the product and we calculate the sum of units sold
quarter4=df4.groupby(['store','Name']).agg({'  u':['sum']})
quarter4.columns = quarter4.columns.droplevel(1)
quarter4 = quarter4.sort_values(by=['store','  u'], ascending=False) #we sort Product names by units sold
quarter4 = quarter4.groupby(level=0).head(5) #we decided to show only the top 5 products for each store based on units sold
```

Here below an example of the results of the groupby: for stores 410 and 409 we can see the top 5 products sold per quarter taking into consideration units sold

```python
quarter1=quarter1.reset_index()
quarter2=quarter2.reset_index()
quarter3=quarter3.reset_index()
quarter4=quarter4.reset_index()
```

```python
#as the store repeats 4 times, in order to merge the different quarters we need first to create a unique index.
#we will use the ranking from 1st to 5th place to differentiate the stores
a = [1,2,3,4,5]
b = a*410
quarter1['ranking']=b

b1 = a*409 #quarter2-3-4 miss 1 store, so we are creating another list repeteated 409 times
quarter2['ranking']=b1
quarter3['ranking']=b1
quarter4['ranking']=b1

quarter1['Store_ranking']= quarter1['store'].astype(str)+ '.' + quarter1['ranking'].astype(str)
quarter2['Store_ranking']= quarter2['store'].astype(str)+ '.' + quarter2['ranking'].astype(str)
quarter3['Store_ranking']= quarter3['store'].astype(str)+ '.' + quarter3['ranking'].astype(str)
quarter4['Store_ranking']= quarter4['store'].astype(str)+ '.' + quarter4['ranking'].astype(str)
```

```python
quarter1.drop(columns = ['store', 'ranking'], inplace = True)
quarter2.drop(columns = ['store', 'ranking'], inplace = True)
quarter3.drop(columns = ['store', 'ranking'], inplace = True)
quarter4.drop(columns = ['store', 'ranking'], inplace = True)
```

```python
topproducts19=quarter1.merge(quarter2,how='outer', on='Store_ranking', suffixes=('_Q1','_Q2'))
topproducts19=topproducts19.merge(quarter3,how='outer', on='Store_ranking')
topproducts19=topproducts19.merge(quarter4, how='outer', on='Store_ranking')
```

```python
topproducts19 = topproducts19[['Store_ranking','Name_Q1', '  u_Q1', 'Name_Q2', '  u_Q2', 'Name_x', '  u_x',
       'Name_y', '  u_y']]
```

```python
topproducts19.columns = ['Store ranking', 'Name_Q1', 'units_Q1', 'Name_Q2', 'units_Q2', 'Name_Q3', 'units_Q3',
       'Name_Q4', 'units_Q4']
```

## Market share
## Families

For the market share, we will check the amount of units sold for each family for each quarter, and we will calculate the percentage of each family wrt the total units sold for that quarter.

```python
df19.head()
```

```python
df=df19
```

```python
df['week'] = pd.to_datetime(df['week'])
df['quarter'] = df['week'].dt.quarter
```

```python
df1=df[df['quarter']==1]
df2=df[df['quarter']==2]
df3=df[df['quarter']==3]
df4=df[df['quarter']==4]
```

```python
# we will use a groupby to show for each store each family and the percentage of units sold
quarter1 = df1.groupby(['store','Family']).agg({'  u':['sum']})
quarter1 = quarter1.groupby(level=0).apply(lambda x:x / float(x.sum())).round(2)
quarter1.columns = quarter1.columns.droplevel(1)
```

```python
quarter2 = df2.groupby(['store','Family']).agg({'  u':['sum']})
quarter2 = quarter2.groupby(level=0).apply(lambda x:x / float(x.sum())).round(2)
quarter2.columns = quarter2.columns.droplevel(1)
```

```python
quarter3 = df3.groupby(['store','Family']).agg({'  u':['sum']})
quarter3 = quarter3.groupby(level=0).apply(lambda x: x / float(x.sum())).round(2)
quarter3.columns = quarter3.columns.droplevel(1)
```

```python
quarter4 = df4.groupby(['store','Family']).agg({'  u':['sum']})
quarter4 = quarter4.groupby(level=0).apply(lambda x: x / float(x.sum())).round(2)
quarter4.columns = quarter4.columns.droplevel(1)
```

```python
quarter1=quarter1.reset_index()
quarter2=quarter2.reset_index()
quarter3=quarter3.reset_index()
quarter4=quarter4.reset_index()
```

```python
#as the store repeats 21 times as the number of families, in order to merge the different quarters 
#we need first to create a unique index.

quarter1['Store_family']= quarter1['store'].astype(str)+ '.' + quarter1['Family'].astype(str)
quarter2['Store_family']= quarter2['store'].astype(str)+ '.' + quarter2['Family'].astype(str)
quarter3['Store_family']= quarter3['store'].astype(str)+ '.' + quarter3['Family'].astype(str)
quarter4['Store_family']= quarter4['store'].astype(str)+ '.' + quarter4['Family'].astype(str)
```

```python
quarter2.drop(columns = ['store', 'Family'], inplace = True)
quarter3.drop(columns = ['store', 'Family'], inplace = True)
quarter4.drop(columns = ['store', 'Family'], inplace = True)
```

```python
marketshare19=quarter1.merge(quarter2,how='outer', on='Store_family', suffixes=('_Q1','_Q2'))
marketshare19=marketshare19.merge(quarter3,how='outer', on='Store_family')
marketshare19=marketshare19.merge(quarter4, how='outer', on='Store_family')
marketshare19.drop(columns = ['Store_family'], inplace = True)
```

```python
#change columns name 
marketshare19.columns = ['store', 'Family', '%units_sold_Q1', '%units_sold_Q2', '%units_sold_Q3', '%units_sold_Q4']
```

```python
marketshare19
```
