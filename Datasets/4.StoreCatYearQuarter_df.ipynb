```python
import dask.dataframe as dd
import datetime
from collections import defaultdict
import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_17 = pd.read_csv('2017.csv')
df_16 = pd.read_csv('2016.csv')
df_18 = pd.read_csv('2018.csv')
df_19 = pd.read_csv('2019.csv')
```

```python
format='%Y-%m-%d'
df_16.drop(columns=['Brand','Name','sku', 'Family'], inplace = True)
df_16['date'] = pd.to_datetime(df_16['week'], format='%Y-%m-%d')
df_16['quarter'] = df_16['date'].dt.quarter
df_16['year'] = df_16['date'].dt.year
df_16.drop(columns=['date','week'], inplace = True)
g16 = df_16.groupby(['store','Category','year','quarter']).sum()
g16.reset_index(inplace = True)
```
